Matplotlib
it's a python package for visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt# This is the way you can import matplotlib
```

```python
arr = np.array([12,13,10,14,16])
```

```python
plt.plot(arr)#By default it will create line graph
```

Histograms are used to show distributions of variables while bar charts are used to compare variables. Histograms plot quantitative data with ranges of the data grouped into bins or intervals while bar charts plot categorical data. This post expands on these differences and mentions several other differences between histograms and bar charts.

```python
plt.hist(arr)#It will create histogram
```

```python
arr
```

```python
arr1 = np.array([8,9,10,10,12])
plt.scatter(arr,arr1)#Scatter plot
plt.xlabel('arr')
plt.ylabel('arr1')
plt.show()
```

Matplotlib - Pie Chart


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])#add_axes([left, bottom, width, height])
ax.set_aspect('equal', adjustable='box')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
val = [20,20,50,5,5]
ax.pie(val, labels = langs,autopct='%1.3f%%')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students, labels = langs,autopct='%1.0f%%')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])#add_axes([left, bottom, width, height])
ax.axis('equal')
name =['Sachin','Sehwag','Ganguly','KLRahul','ViratKohli']
avr = [50,20,10,10,10]
ax.pie(avr,labels=name,autopct='%1.2f%%' )

plt.show()
```

```python
#Barchart
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
avr1 = [45,70,45,55,45]
ax.bar(name,avr1)
plt.show()
```

```python
data = [[50, 25, 50, 20],
[45, 23, 51, 17],
[60, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Avrage run in 2020')
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'y', width = 0.25)
ax.legend(labels=['Sachin','Sehwag','Ganguly'])
plt.show()
```
