```python
def get_bin(x):
    if x<10:
        return "<10"
    elif x>=10 and x<20:
        return "10-20"
    elif x >=20 and x< 30:
        return "20-30"
    elif x>=30:
        return "30+"
    else:
        return "Not known"
```

```python
car_sales2["Price_cat2"]=car_sales2["Price_in_thousands"].apply(get_bin)
```
